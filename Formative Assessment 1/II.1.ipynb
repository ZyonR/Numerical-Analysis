Question 2.1: Sketch the graphs of y = $e^x-2$ and y = cos($e^x-2$). Use the Bisection method to find an  approximation to within$ 1^{0âˆ’}$5 to a value in$ [0.5,1.5$] with$e^x-2$2 = cos$e^x-2$2)

```python
def bisectionMethod(a, b, function, tolerance):
    a_n__b_n = []
    p_n = []
    f_xn = []

    dist = b - a
    while dist >= tolerance:
        a_n__b_n.append((a, b))
        p = (a + b) / 2
        p_n.append(p)
        f_p = function(p)
        f_xn.append(f_p)

        if function(a) * f_p < 0:
            b = p
        else:
            a = p

        dist = b - a

    df = {
        "a_n": [interval[0] for interval in a_n__b_n],
        "b_n": [interval[1] for interval in a_n__b_n],
        "p_n": p_n,
        "f_xn": f_xn
    }
    bisected = pd.DataFrame(df)
    return bisected
```

```python
def func(x):
    v = (m.e**x)-2
    return m.cos(v)-v
```

```python
bisected = bisectionMethod(0.5,1.5,func,pow(10,-5))
bisected
```
