Question 2: Use the Bisection method to find solutions accurate to within $10^{-2}$ for $x^3 - 7x^2 + 14x - 6 = 0$

```python
def bisectionMethod(a, b, function, tolerance):
    a_n__b_n = []
    p_n = []
    f_xn = []

    dist = b - a
    while dist >= tolerance:
        a_n__b_n.append((a, b))
        p = (a + b) / 2
        p_n.append(p)
        f_p = function(p)
        f_xn.append(f_p)

        if function(a) * f_p < 0:
            b = p
        else:
            a = p

        dist = b - a

    df = {
        "a_n": [interval[0] for interval in a_n__b_n],
        "b_n": [interval[1] for interval in a_n__b_n],
        "p_n": p_n,
        "f_xn": f_xn
    }
    bisected = pd.DataFrame(df)
    return bisected
```

```python
def func(x):
    return (x**3)-7*(x**2)+14*x-6
```

```python
bisected = bisectionMethod(0,1,func,pow(10,-2))
bisected
```

```python
bisected = bisectionMethod(2.8,3.3,func,pow(10,-2))
bisected
```

```python
bisected = bisectionMethod(3.3,4,func,pow(10,-2))
bisected
```
