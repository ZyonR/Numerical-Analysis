Question 6: Determine an interval [a,b] on which fixed-point iteration will converge for $x= \frac{5}{x^2}+2$. Estimate the number of iterations necessary to obtain approximations accurate to within $10^{-5}$.

```python
import math as m
import pandas as pd

def fixedPointMethod(p0,tolerance,gX):
    pn = []
    gn = []
    while abs(p0 - gX(p0)) >= tolerance:
        gn.append(gX(p0))
        pn.append(p0)
        p0 = gX(p0)
    fpm = {
        "g_n":gn,
        "p_n":pn,
    }
    return pd.DataFrame(fpm)

def gX(p):
    return 5*m.pow(p,-2)+2
```

```python
fixed = fixedPointMethod(2.25,m.pow(10,-5),gX)
fixed
```
